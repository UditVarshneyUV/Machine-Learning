```python
# Fitting Polynomial Regression to the dataset with degree=4

poly_reg = PolynomialFeatures(degree = 4)
X_poly = poly_reg.fit_transform(X)
reg2 = LinearRegression()
reg2.fit(X_poly, Y)

#Prediction Value
Y_pred = reg2.predict(poly_reg.fit_transform([[4]]))
print([Y_pred,Y[4]])

# Visualising the Polynomial Regression results with degree=4
X_grid = np.arange(min(X), max(X), 0.1)
X_grid = X_grid.reshape((len(X_grid), 1))
plt.scatter(X, Y, color='red')
plt.plot(X_grid, reg2.predict(poly_reg.fit_transform(X_grid)), color='blue')
plt.title('Truth or Bluff (Linear Regression)')
plt.xlabel('Position')
plt.ylabel('Salary')
plt.show()
```
